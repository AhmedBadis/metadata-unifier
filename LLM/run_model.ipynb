```python
import subprocess

# Build the Docker image
build_command = ["docker", "build", "-t", "chainlit-llm", "."]
try:
    subprocess.run(build_command, check=True, capture_output=True, text=True)
except subprocess.CalledProcessError as e:
    print("Error during Docker build:")
    print(e.stdout)
    print(e.stderr)

# Run the Docker container
run_command = [
    "docker", "run", "--gpus", "all", "-p", "8501:8501", "--network", "host", "chainlit-llm"
]
subprocess.run(run_command, check=True)
```
